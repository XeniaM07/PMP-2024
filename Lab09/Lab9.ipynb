```python
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

theta_values = [0.2, 0.5]
y_values = [0, 5, 10]

fig, axes = plt.subplots(len(theta_values), len(y_values), 
                         figsize=(18, 8),
                         constrained_layout=True)

for i, theta in enumerate(theta_values):
    for j, y in enumerate(y_values):
        with pm.Model() as model:
            n = pm.Poisson("n", mu=10)
            y_obs = pm.Binomial("y_obs", n=n, p=theta, observed=y)
            trace = pm.sample(1000, return_inferencedata=True, progressbar=False)
            
        az.plot_posterior(trace, var_names=["n"], hdi_prob=0.94, ax=axes[i, j])
        axes[i, j].set_title(f"O={theta}, Y={y}", fontsize=10)

fig.suptitle("Distribuția a posteriori a lui n pentru diferite combinații de O și Y", fontsize=16)
plt.show()
```
